```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x  = np.linspace(10,20,100)
plt.plot(x,np.sin(x),'g-')
plt.plot(x,np.cos(x),'r.')
plt.xlabel('angle')
plt.ylabel('values')
plt.title('Graph')
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax  = fig.add_axes([0,0,1,1])
b = ["MP","UP","RJ"]
c = [100,500,300]
ax.bar(b,c)
plt.xlabel("State")
plt.ylabel("Count")
plt.show()
```

```python
fig = plt.figure()
ax  = fig.add_axes([0,0,1,1])
b = ["MP","UP","RJ"]
c = [100,500,300]
ax.pie(c,labels = b,autopct =' %0.1f%%')

plt.show()
```

```python
import numpy as np
z = 20*np.random.random(10000)
s = np.sin(z)
d = np.cos(z)
fig = plt.figure()
ax = plt.axes(projection ='3d')
ax.scatter(s,d,z,c=z,cmap ='Greens')
plt.show()
```
